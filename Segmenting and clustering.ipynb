## Week3: Segmenting and Clustering

#### Part 1 Create neighbourhood dataframe

```python
#import pandas
import pandas as pd
```

```python
#read data from website
PosCode=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
PosCode=PosCode[0]
```

```python
#make the first row of dataframe as column names
PosCode.columns=PosCode.iloc[0]
PosCode=PosCode.drop(PosCode.index[0])
PosCode
```

```python
#Remove 'Not assigned' Borough
Pos_assigned=PosCode[PosCode.Borough!="Not assigned"]
```

```python
#Replace 'Not assigned' neighbourhood same as its borough
Pos_assigned.loc[Pos_assigned['Neighborhood'] == ('Not assigned'), 'Neighborhood'] = Pos_assigned['Borough']
Pos_assigned
```

```python
#Group neighbourhoods with same postal code
Pos_merged=Pos_assigned.groupby(['Postcode','Borough'],as_index=True,sort=False).agg(lambda x:','.join(x)).reset_index()
Pos_merged
```
