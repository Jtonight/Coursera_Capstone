## Week3: Segmenting and Clustering

#### Part 1 Create neighbourhood dataframe

```python
#import pandas
import pandas as pd
```

```python
#read data from website
PosCode=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
PosCode=PosCode[0]
```

```python
PosCode
```

```python
#Remove 'Not assigned' Borough
Pos_assigned=PosCode[PosCode.Borough!="Not assigned"]
```

```python
#Replace 'Not assigned' neighbourhood same as its borough
Pos_assigned.loc[Pos_assigned['Neighborhood'] == ('Not assigned'), 'Neighborhood'] = Pos_assigned['Borough']
Pos_assigned
```

```python
#Group neighbourhoods with same postal code
Pos_merged=Pos_assigned.groupby(['Postcode','Borough'],as_index=True,sort=False).agg(lambda x:','.join(x)).reset_index()
Pos_merged
```

```python
Pos_merged.shape
```

#### Part 2 Assigning coordinates

```python
#read geospatial data
geo=pd.read_csv('C:/Python/Geospatial_Coordinates.csv')
geo.rename(columns={'Postal Code':'Postcode'},inplace=True)
geo
```

```python
df=pd.merge(Pos_merged,geo,on=['Postcode'],how='outer')
df
```

#### Part 3 Segmenting and Clustering

```python
import folium # map rendering library
!conda install -c conda-forge geopy --yes 
from geopy.geocoders import Nominatim 
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="Toronto")
location = geolocator.geocode(address)
latitude= location.latitude
longitude= location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'],df['Longitude'], df['Borough'], df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
